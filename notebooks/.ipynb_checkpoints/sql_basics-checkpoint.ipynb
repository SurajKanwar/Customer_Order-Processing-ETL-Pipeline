# Day 2: SQL Fundamentals

```python
import pandas as pd
import mysql.connector
```

```python
csv_file = r"C:\Users\Dusty\Downloads\Internship\Git_Repo\Customer-Order-Processing-ETL-Pipeline\data\sample_orders.csv"
df = pd.read_csv(csv_file)

# Establish connection to the MySQL database
conn = mysql.connector.connect(
    host="localhost",        # Hostname (or IP address)
    user="Suraj",    # MySQL username
    password="root", # MySQL password
    database="sample_orders" # Database name
)
# Create a cursor object
cursor = conn.cursor()

# Create the 'orders' table if it doesn't exist
cursor.execute("""
    CREATE TABLE IF NOT EXISTS orders (
        order_id INT PRIMARY KEY,
        customer_name VARCHAR(255),
        product_name VARCHAR(255),
        quantity INT,
        price DECIMAL(10, 2),
        order_date DATE
    );
""")

# Insert data into MySQL
for _, row in df.iterrows():
    cursor.execute("""
        INSERT INTO orders (order_id, customer_name, product_name, quantity, price, order_date)
        VALUES (%s, %s, %s, %s, %s, %s)
    """, (row['order_id'], row['customer_name'], row['product_name'], row['quantity'], row['price'], row['order_date']))

# Execute a query
cursor.execute("select * from orders;")
rows = cursor.fetchall()

for row in rows:
    print(row)

# Close the connection
conn.commit()
cursor.close()
conn.close()
```
