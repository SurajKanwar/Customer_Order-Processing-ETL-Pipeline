```python
import pandas as pd
```

```python
df=pd.read_csv(r'C:\Users\Dusty\Downloads\Internship\Git_Repo\Customer-Order-Processing-ETL-Pipeline\data\sample_orders.csv', header='infer')
```

```python
df.head()
```

```python
df ["price"] = df ["price"].round(2)
df.head()
```

```python
df.info()
```

```python
df["customer_name"] = df["customer_name"].astype('string')
df["product_name"] = df["product_name"].astype('string')
df['order_date'] = pd.to_datetime(df['order_date'])
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df=df.drop_duplicates()
```

```python
df
```

```python
df['price'] = df.price.fillna(df.price.mean())
```

```python
df["TotalPrice"] = df['quantity'] * df['price']
df
```

```python
df.rename(columns= { 'order_id' : 'OrderID', 'customer_name': 'CustomerName', 'product_name':'ProductName',
                     'order_date':'OrderDate', 'quantity': 'Quantity', 'price': 'Price'},inplace=True)
```

```python
df
```
